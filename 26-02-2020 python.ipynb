### File handling:
* by using file handling we can write or read or append in python
* performing operations on a file is called as file handling
* entering some data to a file is called file writing
* retriving data from a file is called file retriving
* we can perform append,delete operators on file

```python
def writetoafile(filename,data):
    with open(filename,"w")as f:
        f.write(data)
    return
filename = "data.txt"
writetoafile(filename,"gudiyala\nnaga\nsai\nram\n")
```

```python
def readfromafile(filename):
    with open(filename,"r") as f:
        fd = f.read()
        f.close()
    return fd
readfromafile("data.txt")
```
